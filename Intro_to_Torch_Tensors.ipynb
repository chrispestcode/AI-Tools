Shape is a crucial property of tensors. Most useful results from tensors are reliant on tensor operations, which are matrix operations in N-dimensional space.



Some background in linear algebra, the study of linear transformations on vector spaces, may be required to have a comprehensive understanding of why shape is so important.


In short, some rule of dimensionality must be obeyed between any two vectors to perform certain tensor operations.


---



````python
x = torch.rand(size=[2,7])
y = torch.rand(size = [3,6])

torch.matmul(x,y)
````



---






The code above ran into an error because the two matrices, x and y, cannot be multiplied together due to their incompatible dimensions.

There are two types of multiplication when it comes to tensors, element-wise multiplication and matrix multiplication.


---






For element-wise multiplication, the shape of the factor matrices must be the same.

Example:

Matrix A:  3 x 2

Matrix B:  1 x 5

Matrix C:  2 x 2


* A * B cannot be done.

* A * C can be done.

* C * A can be done.

The product matrix will be the same size as the factor matrices.

---

````python
A = torch.rand(size = [3, 2])
B = torch.rand(size = [1, 5])
C = torch.rand(size = [3, 2])
A, B, C
````

````python
A*B
````

````python
A * C
````

````python
C * A
````



---



Matrix multiplication differs from element-wise multiplication in that it must obey two different rules, and it sums the values at the product of the ith row of matrix A and ith column of matrix B.


Two rules of matrix multiplcation:

1) The inner dimensions must match.
    
    OK (3, 2) @ (2, 2)

    NOT (3, 1) @ (3, 6)

2) The resulting matrix will have the shape of the outer dimensions of the product matrices.

    (6, 2) @ (2, 9) = (6, 9)

See the code snippets below and note these outcomes:
D @ E cannot be done.

D @ F can be done.

F @ D cannot be done.



---



````python
D = torch.rand(size = [3, 2])
E = torch.rand(size = [1, 5])
F = torch.rand(size = [2, 4])
D, E, F
````

````python
D @ E
````

````python
D @ F
````

````python
F @ D
````

Transposition of a matrix is matrix operation that is used to swap the dimensions and values of a tensor.

If A is 3x2 tensor, it's transpose will be a 2x3 tensor, while swapping its values along its diagonal.

An error was raised before when matrix multiplication was applied to two tensors of different dimensions, D and F with dimensions 3x2 and 2x4, respectively. Performing matrix multiplication using transpositions would make them compatible for matrix multiplication.

Matrix operations can be used in combination to perform linear transformations for use in neural networks, specifically in the feed-forward process of training a layer of network.

In a feed-forward network architecture, a neuron within a layer will accept input, perform a  transformation on it, then return an output. It is important to mention that this transformation could be linear or *non-linear*.

For context, introducing a linear transformation could be simply applying weights and adding biases to inputs on a neuron, or even extract specific values, such as multiplication by an identity matrix. On the other hand, introducing non-linear transformations could be purposeful in confining the range of output values, such as in use in probability, and providing better fit for the solution, in problem spaces where the solution requires non-linearity, such as complex classification problems.

`.view()` does not create a new tensor like `.reshape()`, but it does create a new view of it the input tensor as a different shape. It is key to note that the tensor outputted by `.view()` still points to the original tensor in memory, so the changes are identical.


Passing a value of `-1` causes the dimension to be set automatically for that index.

---



````python
fixed_tensor = torch.arange(start=0., end=100., step=10)
new_tensor = fixed_tensor.view((5,2))
new_tensor
````

````python
new_tensor[0][1] = 30
fixed_tensor
````

Compatability of data type is as important as compatability of shape. Data types for tensors represent numerical values, and are distinguishable based on the floating point precision.

Conversion from a more precise floating point (i.e. 64-bit floating point, float64) to a less precise floating point (i.e. 32-bit floating point, float32) would result in a loss of information for the value and a increase in computation speed.

However, going from a lesser precision point to a higher precision point would result in an increase in calculation precision and slower computational speed.

Data types can be converted using the `.type()` class method of a tensor to create a new tensor with the same values as the original tensor.



---



````python
tensor = torch.arange(10, dtype=torch.float32)
print(tensor)
print(tensor.dtype)

tensor.type(torch.float64)
raise Exception("Print the new data type. Did it change?")
````

It should be noted that calling the method on the tensor is not enough to convert the data type. The `.type()` method creates a *new* instance of the same tensor with the desired data type.

````python
new_tensor = tensor.type(torch.float64)
new_tensor.dtype

````

````python
# You can also instantiate tensors of a desired data type
tensor = torch.tensor([[2, 18]], dtype=torch.float64)
tensor.dtype
````



---



The default data type for tensors instantiated in torch is float32.

It is important to note that this is not the default for all libraries. For example, the default data type for tensors instantiated in numpy is float64.

NumPy is a popular python library for mathematical computations on N dimensional arrays. For the most part, the methods introduced here for PyTorch are nearly identical for NumPy.

PyTorch is optimized for GPU computations, NumPy is optimized for CPU computations. Thus, converting and moving tensors between CPU and GPU is a common strategy to optimize for resource use and speed.

Conversion from **PyTorch tensors to NumPy arrays** can be done using the `.numpy()` class method of a tensor.

Conversion from **NumPy arrays to PyTorch tensors** can be done using the `torch.from_numpy()` method with the numpy array as a parameter.




````python
import numpy as np

tensor = torch.arange(10)
print(f'PyTorch tensor: {tensor}')

# To convert from PyTorch tensors to NumPy arrays
tensor_to_array = tensor.numpy()
print(f'PyTorch tensor to NumPy array: {tensor_to_array}')

# To convert from NumPy arrays to PyTorch tensors
array_to_tensor = torch.from_numpy(tensor_to_array)
print(f'NumPy array to PyTorch tensor: {array_to_tensor}')
````

Moving between CPU and GPU requires access to a GPU. Majority of computers have built-in CPU as their only processing unit.


It is best to run a device check to see whether it is available, and set a device variable for transferring later.


````python
device = "cuda" if torch.cuda.is_available() else "cpu"
device
````

Apple's Silicon M1/M2/M3 GPUs use the `torch.backends.mps` module instead to check for available. If the device does not have these GPUs or run on a Mac device, it will return `false`.

````python
device = "mps" if torch.backends.mps.is_available() else "cpu"
device
````

A strategy approach can be used to check for all of these cases at once.

````python
if torch.cuda.is_available():
    device = "cuda" # Use NVIDIA GPU (if available)
elif torch.backends.mps.is_available():
    device = "mps" # Use Apple Silicon GPU (if available)
else:
    device = "cpu" # Default to CPU if no GPU is available
device
````

To move tensors and arrays between CPU and GPU, use the `.to()` and `.cpu()`.

`.to()` moves tensors to GPU.

`.cpu()` moves tensors to CPU.

Calling `.numpy()` on a tensor in GPU will throw an *error*. It is necessary to move tensors from GPU to CPU, then transform them into NumPy arrays.






````python
tensor = torch.ones((3,4))
tensor_on_gpu = tensor.to(device)
tensor_on_gpu # Google Colab is set to cpu, so the output will look normal
````

````python
tensor_on_cpu = tensor_on_gpu.cpu()
tensor_cpu_as_numpy = tensor_on_cpu.numpy()
tensor_on_cpu, tensor_cpu_as_numpy
````

It is important to note the changing values on the numpy array changes the values of the cpu tensor if they share memory. This also affects the "gpu" tensor if it also on cpu, as is the case of no available GPUs.

To work around this, use `.copy()` to create a copy of the tensor to ensure it does not share memory.

````python
tensor_cpu_as_numpy[0][1] = 20
tensor_on_cpu, tensor_cpu_as_numpy, tensor_on_gpu
````

````python
tensor_on_cpu_copy = tensor_cpu_as_numpy.copy()
tensor_on_cpu_copy[0][1] = 515
tensor_on_cpu, tensor_cpu_as_numpy, tensor_on_gpu, tensor_on_cpu_copy
````

Each tensor contains data that can be processed to reveal aggregational information like the max and min values. PyTorch has built-methods to achieve certain aggregations.

These include:
  * max()
  * min()
  * sum()
  * mean()
  * argmax() # returns the index of the max value
  * argmin() # returns the index of the min value

````python
tensor = torch.arange(start=1, end = 12, step = 1, dtype=torch.float32)
print(tensor)
print(f'min: {tensor.min()}')
print(f'max: {tensor.max()}')
print(f'sum: {tensor.sum()}')
print(f'mean: {tensor.mean()}')
print(f'argmax: {tensor.argmax()}')
print(f'argmin: {tensor.argmin()}')
````
